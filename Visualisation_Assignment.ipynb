##**5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.**

````python
import seaborn as sns
import numpy as np
import pandas as pd


np.random.seed(0)
n_samples = 100
n_features = 5

cov_matrix = np.array([[1, 0.8, 0.2, 0.1, 0.5],
                       [0.8, 1, 0.7, 0.3, 0.2],
                       [0.2, 0.7, 1, 0.9, 0.4],
                       [0.1, 0.3, 0.9, 1, 0.6],
                       [0.5, 0.2, 0.4, 0.6, 1]])

data = np.random.multivariate_normal(mean=np.zeros(n_features), cov=cov_matrix, size=n_samples)
df = pd.DataFrame(data, columns=[f'Feature {i}' for i in range(n_features)])

# Visualize the correlation matrix of a dataset using a heatmap
sns.set()
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.show()
````

````python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}


df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='X', y='Y', z='Z', title='3D Scatter Plot of Random Data')
fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)


fig.show()
````

````python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(df, x="Grade", y="Score", box=True, points="all")
fig.update_layout(title='Violin Plot of Scores by Grade',
                  xaxis_title='Grade',
                  yaxis_title='Score')


fig.show()
````

##**3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.**
```
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
    }
df = pd.DataFrame(data)
```

````python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(values='Sales', index='Month', columns='Day', aggfunc='sum', fill_value=0)

fig = px.imshow(heatmap_data,
                labels=dict(x="Day", y="Month", color="Sales"),
                title="Heatmap of Sales Variation Across Months and Days",
                color_continuous_scale='Viridis')


fig.show()
````

##**4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(√(x²+y²))**

```
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()

```

````python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)


fig = go.Figure(data=[go.Surface(z=z, x=x[0], y=y[:,0], colorscale='Viridis')])
fig.update_layout(title='3D Surface Plot of z = sin(√(x² + y²))',
                  scene=dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'),
                  autosize=True)

fig.show()

````

##**1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x**



````python

from bokeh.plotting import figure, show, output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 200)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)',
           width=800, height=400)

p.line(x, y, line_width=2, color="navy")

show(p)
````

##**2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns**

````python

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

output_notebook()

np.random.seed(42)
num_points = 100

x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 10
sizes = np.random.randint(5, 20, num_points)
colors = np.random.choice(['red', 'blue', 'green', 'yellow', 'purple'], num_points)

data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})


source = ColumnDataSource(data)

p = figure(title="Scatter Plot with Random Sizes and Colors",
           x_axis_label='x', y_axis_label='y', width=800, height=400)


p.scatter(x='x', y='y', size='sizes', color='colors', source=source,
          alpha=0.6)

show(p)
````

##**3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**
```
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

```

````python

from bokeh.plotting import figure, show, output_notebook

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts",
           toolbar_location=None, tools="", width=800, height=400)

p.vbar(x=fruits, top=counts, width=0.9, color="navy")

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
````

##**4. Create a Bokeh histogram to visualize the distribution of the given data.**
```
data_hist = np.random.randn(1000)

```

````python

from bokeh.plotting import figure, show, output_notebook
import numpy as np

output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Data', y_axis_label='Frequency',
           width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

p.y_range.start = 0

show(p)
````

##**5. Create a Bokeh heatmap using the provided dataset**
```
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

````python

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y',
           width=800, height=400, tools="pan,wheel_zoom,box_zoom,reset,save",
           x_range=[0, 1], y_range=[0, 1])

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
````
